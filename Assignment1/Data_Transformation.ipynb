```python
import numpy as np
import pandas as pd
lst={'Name':['Ram','Sham','Geeta','Sita','XX','YY','ZZ'],'Age':[14,3,12,20,21,60,65],'Gender':['M','M','F','F','M','M',np.nan]}
df1=pd.DataFrame(lst)
print(df1)
```

```python
#Find the categorical column and separate it out with a different dataframe. To do so, use the select_dtypes()
df_cat = df1.select_dtypes(exclude=[np.number])
print(df_cat)
```

```python
#Find the distinct unique values in the Grade column using unique() function
df_cat['Gender'].unique()
```

```python
#Find the frequency distribution of each categorical column using value_counts()
df_cat.Gender.value_counts()
```

```python
#Replace the entries in the Gender column. Replace Male with 0 and Female with 1. To do so, use the replace() function:
df_cat.replace(['M','F'],[1,0])
```
