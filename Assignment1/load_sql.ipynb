```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
username = "connector"
password = "password"
host = "localhost"
database = "Movie"

engine = create_engine(
    f"mysql+mysqlconnector://{username}:{password}@{host}/{database}"
)
```

```python
query = "SELECT * FROM Movie"
```

```python
df = pd.read_sql(query, engine)
```
