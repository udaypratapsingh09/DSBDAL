# Part 1
Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical) variable. For example, if your categorical variable is age groups and quantitative variable is income, then provide summary statistics of income grouped by the age groups. Create a list that contains a numeric value for each response to the categorical variable. 

```python
df = pd.read_csv("Salary_Data.csv")
df.head()
```

```python
education_stats = df.groupby('Education Level')['Salary'].agg(['mean', 'median', 'min', 'max', 'std'])

print("Summary Statistics of Age grouped by Income:")
print(education_stats)
```

```python
bins = [0, 5, 10, 15, 20, 80]
# [0-12, 13-18, 19-39, 40-59, 60-80] 
labels = ["0-5", "5-10", "10-15","15-20", "20+"]

df["Experience"] = pd.cut(df["Years of Experience"], bins=bins, labels=labels)

experience_stats = df.groupby("Experience", observed=True)["Salary"].agg(['mean', 'median', 'min', 'max', 'std'])
print(experience_stats)
```

```python
income_bins = [0, 50000, 100000, 150000, 200000, 300000]
income_labels = ["<50k", "50k-100k", "100k-150k", "150k-200k", "200k+"]

df["Income Group"] = pd.cut(df["Salary"], bins = income_bins, labels = income_labels)

age_income = df.groupby("Income Group", observed=True)["Age"].agg(['mean', 'median', 'min', 'max', 'std'])
print(age_income)
```

# Part 2
Write a Python program to display some basic statistical details like percentile, mean, standard deviation of species of ‘Iris-Setosa’, ‘Iris-versicolor’ and ‘Iris-versicolor of iris.csv dataset.

```python
df = pd.read_csv("Iris.csv")
```

```python
# List of species we want to analyze
species_list = set(df["Species"])

for species in species_list:
    print(f"\n--- Statistical Details for {species.upper()} ---")
    
    # Filter the dataframe for the specific species
    species_df = df[df['Species'] == species]
    
    stats = species_df.describe()
    
    print(stats)
```

```python
# Shows the mean for all features, grouped by species
print(df.groupby('Species').mean())

# Shows the standard deviation for all features
print(df.groupby('Species').std())
```
